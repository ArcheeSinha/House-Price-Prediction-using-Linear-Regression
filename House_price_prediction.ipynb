```python
import pandas as pd 
import seaborn as sns
```

```python
house_prices = pd.read_csv("HousePricePrediction.csv")
```

```python
house_prices.shape
house_prices.describe()
house_prices.info()
```

```python
house_prices.head()
```

```python
house_prices.drop(['Id'], axis=1, inplace=True)


house_prices["SalePrice"] = house_prices["SalePrice"].fillna(
    house_prices["SalePrice"].mean()
)

house_prices.dropna(inplace = True)

print(house_prices.isnull().sum())

house_prices.head()
```

```python
house_prices
```

```python
# Data Preprocessing - Encoding 
cols = ['MSZoning', 'LotConfig', 'BldgType', 'Exterior1st']
house_prices = pd.get_dummies(house_prices, columns=cols, drop_first=True)
```

```python
house_prices.head()
```

```python
#Train Test split
from sklearn.model_selection import train_test_split

x = house_prices.drop(["SalePrice"], axis = 1)
y = house_prices["SalePrice"]

x_train, x_test, y_train, y_test = train_test_split(
    x,y, test_size= 0.2, random_state= 42
)
```

```python
x_test.head()
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
# Evaluate 
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print(f"R^2 Score: {r2}")

n = x_test.shape[0]
p = x_test.shape[1]

adjusted_r2 =  1 - ((1-r2)*(n-1)/ (n-p-1))
print("adjusted r^2:", adjusted_r2)
```
